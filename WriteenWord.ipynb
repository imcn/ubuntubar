```python
import docx
from docx.enum.text import WD_ALIGN_PARAGRAPH
import pandas as pd
```

```python
# 导入需要处理的电子表格和文档
file_name = "e:\儿童建档\泸县人民医院医共体卫生进校园工作方案(1015初稿).docx"   # 读取word模版
doc = docx.Document(file_name)
df = pd.read_excel('e:\儿童建档\（泸县惠济路幼儿园）2024年秋季儿童建档信息登记表.xls', sheet_name='Sheet1')
df['班级'] = df['班级'].ffill()
```

```python
for item in range(df.shape[0]):
    content = [list(row) for _, row in df.iterrows()]               # 遍历表中的数据，每行生成一个列表，再组成一个大列表
    table = doc.tables[0]                                           # Word文档中的第一个表格
    index = [[0, 7], [0, 11], [0, 17],                              # Word文档中的空格坐标
             [1, 7], [1, 11], [1, 17]]
    for i in range(0, 6):                                           # 需要填入一行的数据个数
        cell = table.cell(index[i][0], index[i][1])                 # 获取单元格
        cell.text = str(content[item][i])                           # 将行中的数据填写到单元格中
        if cell.text == 'nan':                                      # 如果是空值，则填写 ‘ ’
            cell.text = ' '
        cell.paragraphs[0].alignment = WD_ALIGN_PARAGRAPH.CENTER    # 格式居中
        for par in cell.paragraphs:
            for run in par.runs:
                run.font.size = docx.shared.Pt(10)                  # 字体大小
                run.font.name = '宋体'                              # 字体

    file_name = 'E:\\儿童建档\\生成表格\\' + str(df.iloc[item, 0]) + '.docx'    # 存档
    # file_name = "E:\\儿童建档\\生成表格\\姓名.docx"
    doc.save(file_name)
    print("保存成功！")
```
